```python
cp = pd.read_csv('/content/online_retail.csv')
```

```python
cp.columns
```

```python
cp['InvoiceDate'] = pd.to_datetime(cp['InvoiceDate'])
cp['TotalPrice'] = cp['Quantity'] * cp['UnitPrice']
```

```python
snapshot_date = cp['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = cp.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'nunique',
    'TotalPrice': 'sum'
}).rename(columns={
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalPrice': 'Monetary'
})
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
wcss = []
sil_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=1)
    kmeans.fit(rfm_scaled)
    wcss.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(rfm_scaled, kmeans.labels_))
```

```python
plt.plot(range(2, 11), wcss, marker='o')
plt.title('Elbow Curve')
plt.show()
```

```python
plt.plot(range(2, 11), sil_scores, marker='o', color='green')
plt.title('Silhouette Scores')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=1)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)
```

```python
cluster_summary = rfm.groupby('Cluster').mean()
print(cluster_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Cluster', palette='Set2')
plt.title('RFM Clusters')
plt.show()
```

```python
cp = cp.dropna(subset=['CustomerID'])
```

```python
cp = cp[~cp['InvoiceNo'].str.startswith('C')]
```

```python
cp = cp[(cp['Quantity']>0) & (cp['UnitPrice']>0)]
```

```python
# Binary matrix: 1 if bought, 0 otherwise
pivot_table = cp.pivot_table(index='CustomerID', columns='StockCode', values='Quantity', aggfunc='sum')
pivot_table = pivot_table.fillna(0)
pivot_table = pivot_table.applymap(lambda x: 1 if x > 0 else 0)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

# Transpose: each product becomes a row (Item-Item matrix)
item_similarity = cosine_similarity(pivot_table.T)
item_sim_cp = pd.DataFrame(item_similarity, index=pivot_table.columns, columns=pivot_table.columns)
```

```python
def get_similar_products(stock_code, top_n=5):
    if stock_code not in item_sim_cp.index:
        return "Product not found."

    # Sort by similarity score, exclude the product itself
    similar_items = item_sim_cp[stock_code].sort_values(ascending=False)
    top_similar = similar_items.iloc[1:top_n+1]  # Skip self (index 0)
    return top_similar
```

```python
# Create a StockCode to Description map
desc_map = cp[['StockCode', 'Description']].drop_duplicates().set_index('StockCode')['Description'].to_dict()

def print_similar_descriptions(stock_code):
    similar = get_similar_products(stock_code)
    if isinstance(similar, str):
        print(similar)
        return
    for code, score in similar.items():
        print(f"{desc_map.get(code, code)} (Code: {code})  Similarity: {score:.2f}")
```

```python
print_similar_descriptions('85123A')  # Suppose '85123A' is a product's StockCode
```
